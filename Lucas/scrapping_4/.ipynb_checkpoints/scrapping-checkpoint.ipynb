```python
liste=[]
with open(r'C:\Users\Lucas\Documents/M2/S2/Big data/VG_Shiny/Lucas/scrapping_3/vgsales.csv', 'r') as data:
    for line in csv.DictReader(data):
        liste.append(line)

no_threads = 16
with concurrent.futures.ThreadPoolExecutor(max_workers=no_threads) as executor:
    for dictionnaire in tqdm(liste):
        executor.submit(webscrapping.new_dict, dictionnaire)
```
